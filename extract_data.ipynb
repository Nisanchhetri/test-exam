```python
garbage_list1 = ['A',')\xa0', 'Strikeout option A', 
                 'Un-strikeout A option', 
                  'B',
 ')\xa0',
 '',
 'Option is eliminated.',
 'Strikeout option B',
 'Un-strikeout B option',
 'C',
 '\xa0',
 'Strikeout option C',
 'Un-strikeout C option',
 'D',
 'Hemophilia A',
 'Strikeout option D',
 'Un-strikeout D option',
 'E',
 'Strikeout option E',
 'Un-strikeout E option',
 'F',
 'Option is eliminated.',
 'Strikeout option F',
 'Un-strikeout F option',
 'G',
 'Strikeout option G',
 'Un-strikeout G option',
 'H',
 'Strikeout option H',
 'Un-strikeout H option',
 'I',
 'Strikeout option I',
 'Un-strikeout I option',
 'J',
 'Strikeout option J',
 'Un-strikeout J option',
 'K',
 'Strikeout option K',
 'Un-strikeout K option',
 'L',
 'Strikeout option L',
 'Un-strikeout L option',
 '\t\t\t\t\t\t',
 '\t\t\t\t\t',
 '\t\t',
 '\t\t  ',
 'Rationale:',
 'Correct Answer: E.',]
garbage_list2 = [item for item in garbage_list1 if item not in ('A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L')]
```

```python
img_filter_list = ['checkmark.gif', 'error2.gif']
def get_anslist(df):
    ans_list = []
    ques_list = []
    imgsrc_list = []
    data_len = df.shape[0]
    for i in range(data_len):
        splitted_text = df.text_data[i].split('\nA\n)')
        question = splitted_text[0].split('\xa0\xa0\n\n')[-1].replace('\n', '')
        question = question.replace('\t', '')
        ans_choices = splitted_text[-1].split('Correct Answer:')[0]

        anslist = [a for a in ans_choices.split('\n') if a not in garbage_list1 and not a.startswith('\t\t')]
        if not len(anslist):
            anslist = [a for a in ans_choices.split('\n') if a not in garbage_list2 and not a.startswith('\t\t')]
            anslist.insert(0, 'A')
        ques_list.append(question)
        ans_list.append(anslist)
        
        img_src = ast.literal_eval(df.img_src[i])
        for img in img_src:
            img_name = img.split('/')[-1]
            print(img_name)
            if not img_name.startswith('strike') and img_name not in img_filter_list:
                print('pass_test:', img_name)
                img_value = img
                break
            else:
                img_value = 'no_image'
        imgsrc_list.append(img_value)
    return ques_list, ans_list, imgsrc_list
```

```python
qlist, alist, ilist = get_anslist(df)
```

```python
df_final = pd.DataFrame({'question':qlist, 'answer':alist, 'image_src':ilist})
```
